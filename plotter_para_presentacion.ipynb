## Bode PreAmp simulado

```python
# Read the text file, using tab as the separator, and specify encoding
df = pd.read_csv('C:/Users/Usuario/OneDrive/Desktop/ITeDA/Presentacion-Informe Avance de Proyecto/Datos Bode/Bode_data_preamp_sin_capacitores.txt', sep="\t", skiprows=1, names=['Frequency', 'V(out_preamp)'], encoding='ISO-8859-1')

# Check the first few rows to understand the structure
print(df.head())
    
# Extract the first element of V(out_fs), which is the dB value
df['Attenuation'] = df['V(out_preamp)'].apply(lambda x: float(x.split(',')[0][1:].replace('dB', '').strip()))  # Clean the dB part

# Plotting the Bode diagram
plt.plot(df['Frequency'], df['Attenuation'], color='red')
plt.xscale('log')
plt.xlim(7e5, 1e10)

# Horizontal line at 17 dB
plt.axhline(y=17, color='blue', linestyle='--', label='17dB')

# Find the intersection point(s)
intersection_indices = np.where(np.diff(np.sign(df['Attenuation'] - 17)))[0]
if len(intersection_indices) > 0:
    freq_intersection = df['Frequency'].iloc[intersection_indices]
    for freq in freq_intersection:
        # Vertical line at intersection point
        plt.axvline(x=freq, color='green', linestyle='--')

        # Annotation
        plt.annotate(f'Freq: {freq:.2e} Hz', 
                     xy=(freq, 17), 
                     xytext=(5, 5), 
                     textcoords='offset points', 
                     arrowprops=dict(facecolor='black', arrowstyle='->'),
                     fontsize=10, color='black', 
                     bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

plt.title('Simulated Bode Diagram of the PreAmplifier', size=15)
plt.xlabel('Frequency (Hz)', size=13)
plt.ylabel('Attenuation (dB)', size=13)
plt.grid()
plt.legend()
plt.savefig('frequency_vs_attenuation_preamp.pdf', dpi=1000)
plt.show()
```
