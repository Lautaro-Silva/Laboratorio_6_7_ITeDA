```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Bode medidicion directa

```python
# Read the CSV file
df3 = pd.read_csv('datos_reales.csv', delimiter=";", names=("Atenuation", "Frequency"))

# Convert columns to numeric values (replace commas with dots for decimal values)
df3['Frequency'] = pd.to_numeric(df3['Frequency'].str.replace(',', '.'), errors='coerce')
df3['Atenuation'] = pd.to_numeric(df3['Atenuation'].str.replace(',', '.'), errors='coerce')

# Drop rows with any NaN values
df3.dropna(inplace=True)

# Print the DataFrame after conversion to check
print("\nDataFrame after conversion:")
print(df3.head())

# Plot only if DataFrame is not empty
if not df3.empty:
    # Scatter plot with switched axes and set the figure size
    ax = df3.plot(kind="scatter", x='Frequency', y='Atenuation', color='red', figsize=(15, 10))

    # Set the title
    plt.title('ScatterPlot of real data of the whole chain', size=20)
    # plt.ylim(10,20)
    # plt.xlim(10**5, 10**7)

    # Show the plot
    plt.grid()
    plt.savefig('diagrama_bode_real_data.png', dpi=1000)
    plt.show()

else:
    print("No numeric data to plot.")
```

## Bode simulados cadena entera

```python
# Read the text file, using tab as the separator, and specify encoding
df = pd.read_csv('Bode_Full_Chain.txt', sep="\t", skiprows=1, names=['Frequency', 'V(out_fs)'], encoding='ISO-8859-1')

# Check the first few rows to understand the structure
print(df.head())

# Extract the first element of V(out_fs), which is the dB value
df['Attenuation'] = df['V(out_fs)'].apply(lambda x: float(x.split(',')[0][1:].replace('dB', '').strip()))  # Clean the dB part

# Plotting
plt.figure(figsize=(10, 10))
plt.plot(df['Frequency'], df['Attenuation'], color='red')
plt.xscale('log')  # Set x-axis to log scale
plt.title('Simulated Bode Diagram of the whole chain', size=15)
plt.xlabel('Frequency (Hz)', size=13)
plt.ylabel('Attenuation (dB)', size=13)
plt.grid()
plt.savefig('frequency_vs_attenuation_whole_chain.png', dpi=1000)
plt.show()
```

## Plot 1PE

```python
# Read the text file, specifying the delimiter as a comma
df = pd.read_csv('Scripts/photon_data_1_photoequivalent.txt', header=None)

# Print the first few rows of the dataframe to check the data
print(df.head())

# Extract the time and amplitude values
t_values = df[0]  # First column
data = df[1]     # Second column

# Plot the resulting function
plt.plot(t_values, data, color='red', label='Output Signal', linewidth=2)
plt.title('Plot of 1 Individual Photon')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (V)')
plt.grid(True)
plt.legend()
plt.savefig('1pe.png', dpi=1000)
```
