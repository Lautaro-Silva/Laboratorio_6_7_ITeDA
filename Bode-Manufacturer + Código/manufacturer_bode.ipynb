```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df3 = pd.read_csv('mag_pahg.csv', names=("frec", "atenuation"))
df3 = df3.tail(-1)
df3 = df3.astype('float')


# scatter plot
df3.plot(kind="line",
         x='frec',
        y='atenuation',
        color='red')
 
# set the title
plt.title('ScatterPlot of PreAmp HighGain')
#plt.xscale('log') 
plt.ylim(32.9,33.1)
plt.xlim(1.1*10**7, 1.5*10**7)

# show the plot
plt.show()
```

```python
# Load your data
df3 = pd.read_csv('mag_pahg.csv', names=("frec", "atenuation"))
df3 = df3.tail(-1)
df3 = df3.astype('float')

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(df3['frec'], df3['atenuation'], color='red', linewidth=2)

# Customize the plot to match previous styles
plt.xscale('log')  # Logarithmic scale for x-axis
plt.grid(True)  # Add grid
plt.title('Bode Diagram of PreAmp HighGain Manufacturer', fontsize=15)
plt.xlabel('Frequency (Hz)', fontsize=13)
plt.ylabel('Attenuation (dB)', fontsize=13)
plt.xlim(100, 10**10)  # Adjust x-axis limits
plt.ylim(df3['atenuation'].min(), df3['atenuation'].max()+1)  # Adjust y-axis limits

# Display the plot
plt.savefig('frequency_vs_attenuation_preamp_fab.pdf', dpi=1000)
plt.show()
```
