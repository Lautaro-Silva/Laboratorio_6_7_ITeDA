```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df3 = pd.read_csv('mag_palg.csv', names=("frec", "atenuation"))
df3 = df3.tail(-1)
df3 = df3.astype('float')


# scatter plot
df3.plot(kind="line",
         x='frec',
        y='atenuation',
        color='red')
 
# set the title
plt.title('ScatterPlot of PreAmp LowGain')
plt.xscale('log') 
plt.ylim(10,20)
plt.xlim(10**5, 10**7)

# show the plot
plt.show()
```

```python
df = pd.read_csv('mag_fasthaper.csv', names=("frec", "atenuation"))
df = df.tail(-1)
df= df.astype('float')
print(df.head())

# scatter plot
df.plot(kind="line",
         x='frec',
        y='atenuation',
        color='red')
 
# set the title
plt.title('ScatterPlot of FastShapper')
plt.xscale('log') 
plt.ylim(20,25)
plt.xlim(1.3*10**7, 6*10**7)

# show the plot
plt.show()
```
